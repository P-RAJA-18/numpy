```python
import numpy as np
```

```python
data = np.genfromtxt('/content/mark.csv.csv',delimiter=',',names=True,dtype=None)
```

```python
print(data)
```

```python
avg_Math=np.mean(data['Math'])
print(avg_Math)
```

```python
total=data['Math']+data['Physics']+data['Chemistry']+data['Computer']+data['English']
print(total)
```

```python
avg_marks=np.mean(total)
print(avg_marks)
```

```python
print(np.max(total))
```

```python
print(np.min(total))
```

```python
print(np.std(total))
```

```python
t=np.argmax(data['Math'])
print(t)
```

```python
t_ind=np.argmax(total)
table=data[:][t_ind]
print(table)
```

```python
total_sorted = np.argsort(-total)
total=np.sort(total)
print(total_sorted)
```

```python
df_sort=data[total_sorted]
df_sort
```
